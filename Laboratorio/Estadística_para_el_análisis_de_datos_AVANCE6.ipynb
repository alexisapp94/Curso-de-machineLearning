```python
# leer la data desde una archivo csv
import pandas as pd  #Sirve para manejar bases de datos 
titanic = pd.read_csv('titanic.csv', delimiter = ",")

titanic_C = titanic[titanic["Embarked"]=="C"]
tc = titanic_C["Embarked"].count()

titanic_S = titanic[titanic["Embarked"]=="S"]
ts = titanic_S["Embarked"].count()

titanic_Q = titanic[titanic["Embarked"]=="Q"]
tq = titanic_Q["Embarked"].count()


print(f"Cantidad Embarque 'C' son: {tc}")
print(f"Cantidad Embarque 'S' son: {ts}")
print(f"Cantidad Embarque 'Q' son: {tq}")
```
